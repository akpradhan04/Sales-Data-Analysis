### Sales Analysis

#### Data Cleanup

```python
import pandas as pd
```

```python
all_data = pd.read_csv("../dataset/Sales_Data.csv")
all_data.head()
```

##### NA Values Cleanup

```python
# Check how many NA values
all_data.isna().sum()
```

```python
# Check the dataset
all_data[all_data.isna().any(axis=1)]

# Drop rows with NA in all the fields
all_data = all_data.dropna(how='all')

# Check how many NA Values
all_data.isna().sum()
```

##### Order Date Column Cleanup

```python
all_data = all_data[~all_data['Order Date'].str.contains('[A-za-z]+') == True]
```

```python
all_data[all_data['Order Date'].str.contains('[A-za-z]+') == True]
```

##### Correct Data Type

```python
all_data.info()
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])

all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data.info()
```

##### Save Cleaned Up Data

```python
all_data.to_csv("../dataset/Sales_Data_Cleaned.csv", index=False)
```
