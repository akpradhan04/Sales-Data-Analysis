# Sales Analysis: Products Sold Together

## Question 4: What items are most frequently bought together?

This analysis identifies product combinations that are frequently purchased in the same order.

**Dataset:** Sales_Data_Cleaned.csv

## Step 1: Import Libraries and Load Data

```python
# Import necessary libraries
import pandas as pd

# Load the sales data
all_data = pd.read_csv("../dataset/Sales_Data_Cleaned.csv")
all_data.head()
```

## Step 2: Identify Duplicate Order IDs

```python
# Find orders with multiple items (duplicated Order IDs)
# keep=False marks all duplicates as True, capturing all orders with multiple products
df = all_data[all_data['Order ID'].duplicated(keep=False)]

# Display sample of orders with multiple items
print(f"Total orders with multiple items: {df['Order ID'].nunique()}")
df.head()
```

## Step 3: Group Products by Order ID

```python
# Create a copy to avoid warnings and group products within each order
df = all_data[all_data['Order ID'].duplicated(keep=False)].copy()

# Combine all products in each order into a single string (comma-separated)
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# Keep only unique order IDs with their grouped products
df[['Order ID', 'Grouped']].drop_duplicates().head(10)
```

## Step 4: Find Most Common Product Pairs

```python
# Import libraries for combinations and counting
from itertools import combinations
from collections import Counter

# Initialize counter
count = Counter()

# Iterate through each grouped order
for row in df['Grouped']:
    # Split the grouped products into a list
    row_list = row.split(',')
    # Generate all possible pairs (combinations of 2) and count them
    count.update(Counter(combinations(row_list, 2)))

# Display top 10 most commonly sold together product pairs
print("=== Top 10 Products Sold Together ===")
for key, value in count.most_common(10):
    print(f"{key[0]} + {key[1]}: {value} times")

# Get top pair for conclusion
top_pair = count.most_common(1)[0]
print(f"\n=== CONCLUSION ===")
print(f"Most Sold Together: {top_pair[0][0]} + {top_pair[0][1]} ({top_pair[1]} times)")
```

## Conclusion

**Answer:** The most commonly sold together products are iPhone and Lightning Charging Cable, followed by Google Phone and USB-C Charging Cable.
