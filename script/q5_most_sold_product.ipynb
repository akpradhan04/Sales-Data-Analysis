# Sales Analysis: Most Sold Product

## Question 5: What product sold the most? Why do you think it sold the most?

This analysis identifies the best-selling product and investigates why it performs so well.

**Dataset:** Sales_Data_Cleaned.csv

## Step 1: Import Libraries and Load Data

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data
all_data = pd.read_csv("../dataset/Sales_Data_Cleaned.csv")
all_data.head()
```

## Step 2: Analyze Products by Quantity Sold

```python
# Group by product and calculate total quantity ordered
product_sales = all_data.groupby('Product').sum()['Quantity Ordered']
print("Products ranked by Quantity Ordered:")
print(product_sales.sort_values(ascending=False))
```

## Step 3: Visualize Product Sales

```python
# Prepare data for plotting
products = [product for product, df in all_data.groupby('Product')]

# Create bar chart showing quantity sold for each product
plt.figure(figsize=(12, 6))
plt.bar(products, all_data.groupby('Product').sum()['Quantity Ordered'], color='teal')

# Add labels and title
plt.xticks(products, rotation='vertical', size=8)
plt.xlabel('Products')
plt.ylabel('Quantity Sold')
plt.title('Total Quantity Sold by Product')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Step 4: Analyze Price Relationship

```python
# Get average price for each product
product_prices = all_data.groupby('Product').mean()['Price Each']
print("Average Price by Product:")
print(product_prices.sort_values(ascending=False))
```

## Step 5: Visualize Price vs Quantity Relationship

```python
# Create dual-axis plot to show relationship between price and quantity sold
fig, ax1 = plt.subplots(figsize=(14, 7))

# Create second y-axis
ax2 = ax1.twinx()

# Bar chart for quantity ordered (left axis)
ax1.bar(products, all_data.groupby('Product').sum()['Quantity Ordered'], 
        color='green', alpha=0.7, label='Quantity Ordered')

# Line plot for price (right axis)
ax2.plot(products, all_data.groupby('Product').mean()['Price Each'], 
         color='blue', marker='o', linewidth=2, markersize=6, label='Price ($)')

# Set labels
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='green')
ax2.set_ylabel('Price ($)', color='blue')
ax1.set_xticks(products)
ax1.set_xticklabels(products, rotation='vertical', size=8)

# Add title
plt.title('Product Quantity Sold vs Price Analysis')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()

# Find top selling product
top_product = product_sales.idxmax()
top_quantity = product_sales.max()
avg_price = product_prices[top_product]
print(f"\n=== CONCLUSION ===")
print(f"Most Sold Product: {top_product}")
print(f"Quantity Sold: {top_quantity:,}")
print(f"Average Price: ${avg_price:.2f}")
```

## Conclusion

**Answer:** The most sold product is the AAA Batteries (4-pack).
